```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import os
import seaborn as sns
print(plt.style.available)
plt.style.use("seaborn-v0_8")

results_root = "resultsCSV"
methods = ['nerf', 'tensorf', 'FDK', 'naf', 'Lineformer']
methods_name = ["NeRF", "TensoRF", "FDK", "NAF", "SAX-NeRF"]
methods = ['nerf', 'tensorf', 'Lineformer']
methods_name = ["NeRF", "TensoRF", "SAX-NeRF"]
num_views = 50

colors = ['blue'] * (len(methods_name)-1) + ['lightcoral']  # Last bar (Ours) in a different color

if __name__ == "__main__":
    # Data
    # psnr_values = [31, 34, 37, 39, 51]  # PSNR values

    ssim_list, psnr_list = [], []
    large_list, small_list, vessel_list, tumor_list = [], [], [], []
    for method in methods:
        df = pd.read_csv(os.path.join(results_root, f"BDMAP_O_{method}_{num_views}.csv"))  
        # df.fillna(0., inplace=True)
        df_p = pd.read_csv(os.path.join(results_root, f"BDMAP_O_{method}_{num_views}_pixel.csv"))  
        # df_p.fillna(0., inplace=True)
        ssim_list.append([df_p["ssim_3d"].tolist()])
        psnr_list.append([df_p["psnr_3d"].tolist()])
        large_list.append([df.iloc[:, 1:5+1].mean(axis=1).tolist()])
        small_list.append([df.iloc[:, 5+1:10+1].mean(axis=1).tolist()])
        vessel_list.append([df.iloc[:, 10+1:15+1].mean(axis=1).tolist()])
        tumor_list.append([df.iloc[:, 15+1:].mean(axis=1).tolist()])
    ssim_data = np.concatenate(ssim_list, axis=0).T
    ssim_m = np.nanmedian(ssim_data, axis=0,)
    psnr_data = np.concatenate(psnr_list, axis=0).T
    psnr_m = np.nanmedian(psnr_data, axis=0,)
    large_data = np.concatenate(large_list, axis=0).T 
    large_m = np.nanmedian(large_data, axis=0,)
    small_data = np.concatenate(small_list, axis=0).T 
    small_m = np.nanmedian(small_data, axis=0,)
    vessel_data = np.concatenate(vessel_list, axis=0).T 
    vessel_m = np.nanmedian(vessel_data, axis=0,)
    tumor_data = np.concatenate(tumor_list, axis=0).T 
    tumor_m = np.nanmedian(tumor_data, axis=0,)

    for idx, (metric_name, metric_data, metric_m) in enumerate(zip(
        ["SSIM", "PSNR", "large organ (DSC)", "small organ (NSD)", "vessel (clDice)", "tumor (NSD)"],
        [ssim_data, psnr_data, large_data, small_data, vessel_data, tumor_data],
        [ssim_m, psnr_m, large_m, small_m, vessel_m, tumor_m]
    )):
        # Create the figure and axis
        fig, ax = plt.subplots(figsize=(3, 3), dpi=150)
        # ax.grid(True, axis="y")

        # Bar chart
        # bars = ax.bar(methods_name, metric_m, color=colors, zorder=2)
        # bars = ax.violinplot(metric_data, showmeans=False, showmedians=True, showextrema=True, quantiles=[[0.25, 0.75]]*len(methods_name), data=methods_name)
        sns.violinplot(metric_data, inner_kws=dict(box_width=5))
        # bars = ax.boxplot(metric_data, tick_labels=methods_name, zorder=3)
        ax.set_xticks([0,1,2], methods_name)

        # Add the annotation with arrow
        x_ours = 4  # Index of "Ours" bar
        y_naF = ssim_m[-2]  # PSNR of NAF (second highest)
        y_ours = ssim_m[-1]  # PSNR of Ours

        # ax.annotate("10.91 dB", 
        #             xy=(x_ours, y_ours), 
        #             xytext=(x_ours, y_naF + 1), 
        #             ha='center', color='red', fontsize=10, #fontweight='bold',
        #             arrowprops=dict(arrowstyle='->', color='red', lw=1.5))

        # Labels and title
        ax.set_title(metric_name)#, fontsize=10)
        ax.set_ylim(0, 100) if "PSNR" not in metric_name else ax.set_ylim(0, 50)
        # ax.set_title("Medicine", fontsize=12, fontweight='bold', loc='left')

        # Formatting
        # ax.spines['top'].set_visible(False)
        # ax.spines['right'].set_visible(False)

        # Show the plot
        plt.tight_layout()
        plt.show()
```
